```python
class Counter:
    def __init__(self, max_value):
        if not isinstance(max_value, int) or max_value < 0:
            raise ValueError("max value must be a non-negative integer.")
        self._max_value = max_value
        self._current_value = 0

    def increment(self):
        if self._current_value < self._max_value:
            self._current_value += 1
        else:
            print(f"Can't increment beyond maximum value of {self._max_value}.")

    def reset(self):
        self._current_value = 0

    def get_value(self):
        return self._current_value

    def get_max_value(self):
        return self._max_value

    def is_at_maximum(self):
        return self._current_value == self._max_value

    def __str__(self):
        return f"Counter(current_value={self._current_value}, max_value={self._max_value})"
```

```python
my_counter = Counter(max_value=3)
print(f"Initial counter: {my_counter}")

my_counter.increment()
my_counter.increment()
my_counter.increment()
print(f"After 3 increments, {my_counter}")

my_counter.increment()
print(f"After trying tempting to increment beyond max, {my_counter}")

my_counter.reset()
print(f"After reset, {my_counter}")

my_counter.increment()
print(f"{my_counter}")
```

```python
class Triangle(Shape):
    def __init__(self, side1, side2, side3, x_coord, y_coord):
        super().__init__(x_coord, y_coord)
        if not all(isinstance(arg, (int, float)) and arg > 0 for arg in [side1, side2, side3]):
            raise ValueError("Sides must be positive numbers.")
        if not (side1 + side2 > side3 and side1 + side3 > side2 and side2 + side3 > side1):
            raise ValueError("The given sides do not form a valid triangle according to the triangle inequality theorem.")

        self._side1 = side1
        self._side2 = side2
        self._side3 = side3

    def get_side1(self):
        return self._side1
    def get_side2(self):
        return self._side2
    def get_side3(self):
        return self._side3

    def calculate_area(self):
        s = self.calculate_perimeter() / 2
        return math.sqrt(s * (s - self._side1) * (s - self._side2) * (s - self._side3))

    def calculate_perimeter(self):
        return self._side1 + self._side2 + self._side3

    def _get_vertices(self):
        V1 = (self._x_coord, self._y_coord)
        V2 = (self._x_coord + self._side3, self._y_coord)

        numerator = self._side1**2 + self._side3**2 - self._side2**2
        denominator = 2 * self._side1 * self._side3
        if denominator == 0:
            raise ValueError("Cannot calculate triangle vertices for a degenerate triangle.")

        cos_alpha = numerator / denominator
        cos_alpha = max(-1.0, min(1.0, cos_alpha))
        alpha = math.acos(cos_alpha)

        V3_x = self._x_coord + self._side1 * math.cos(alpha)
        V3_y = self._y_coord + self._side1 * math.sin(alpha)
        V3 = (V3_x, V3_y)
        return V1, V2, V3

    def get_perimeter_points(self):
        points = []
        V1, V2, V3 = self._get_vertices()

        def _get_segment_points(start_point, end_point, num_points_on_segment):
            segment_points = []
            if num_points_on_segment <= 0:
                return []
            for i in range(num_points_on_segment):
                if num_points_on_segment == 1:
                    t = 0.0
                else:
                    t = i / num_points_on_segment

                x = start_point[0] + t * (end_point[0] - start_point[0])
                y = start_point[1] + t * (end_point[1] - start_point[1])
                segment_points.append((x, y))
            return segment_points

        points.extend(_get_segment_points(V1, V2, 6))
        points.extend(_get_segment_points(V2, V3, 5))
        points.extend(_get_segment_points(V3, V1, 5))

        return points

    def is_inside(self, px, py):
        V1, V2, V3 = self._get_vertices()

        def cross_product_2d(p1, p2, p3):
            return (p2[0] - p1[0]) * (p3[1] - p1[1]) - (p2[1] - p1[1]) * (p3[0] - p1[0])

        cp1 = cross_product_2d(V1, V2, (px, py))
        cp2 = cross_product_2d(V2, V3, (px, py))
        cp3 = cross_product_2d(V3, V1, (px, py))
        return (cp1 >= 0 and cp2 >= 0 and cp3 >= 0) or \
               (cp1 <= 0 and cp2 <= 0 and cp3 <= 0)

    def __str__(self):
        return f"Triangle(Side1={self._side1}, Side2={self._side2}, Side3={self._side3}, Corner=({self._x_coord}, {self._y_coord}))"
```
