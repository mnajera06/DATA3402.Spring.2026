```python
def generate_uniform(N,mymin, mymax):
    if not isinstance(N, int):
        print("N must be an integer")
    if N < 0:
        print("N must be above 0")
    if not isinstance(mymin, int) or not isinstance(mymax, int):
        print("mymin and mymax must be integers.")
    if mymin > mymax:
        print("mymin must be less than mymax")

    return [random.randint(mymin, mymax) for _ in range(N)]
```

```python
generate_uniform(20, 1, 20)
```

```python
## b)
def variance(data):
   n = len(data)
   m = sum(data)
   if n < 2:
        print("Need at least 2 values to calculate variance.")
        return None

   m = sum(data)
   mean_value = m / n
   squared_diffs = [(x - mean_value)**2 for x in data]

   return sum(squared_diffs) / (n - 1)
```

```python
print ("Variance of Data:", variance(data))
```

```python
def histogram(x,n_bins=10,x_min=None,x_max=None):
    if not isinstance(x, list):
        print("x must be a list.")
        return [], []
    if not x:
        print("x cannot be an empty list.")
        return [], []
    if not isinstance(n_bins, int) or n_bins <= 0:
        print("n_bins must be a positive integer.")
        return [], []

    if x_min is None:
        actual_x_min = min(x)
    else:
        actual_x_min = x_min

    if x_max is None:
        actual_x_max = max(x)
    else:
        actual_x_max = x_max

    if actual_x_max <= actual_x_min:

        if all(val == actual_x_min for val in x):
            hist = [len(x)]
            bin_edges = [actual_x_min, actual_x_min + 1.0]
            return hist, bin_edges
        else:
            print("Invalid")
            return [], []

    bin_size = (actual_x_max - actual_x_min) / n_bins
    hist = [0] * n_bins
    bin_edges = [actual_x_min + i * bin_size for i in range(n_bins + 1)]


    for value in x:
        if value < actual_x_min or value > actual_x_max:
            print
        for i in range(n_bins):
            lower_bound = bin_edges[i]
            upper_bound = bin_edges[i+1]

            if i == n_bins - 1:
                if lower_bound <= value <= actual_x_max:
                    hist[i] += 1
                    break
            else:
                if lower_bound <= value < upper_bound:
                    hist[i] += 1
                    break
    return hist, bin_edges
```

```python

def draw_histogram(x,n_bins,x_min=None,x_max=None,character="#",max_character_per_line=20):
    hist, bin_edges = histogram(x, n_bins, x_min, x_max)
    if not hist:
        return [], []

    max_count = max(hist)
    if max_count == 0:
        print("No data to display ")
        return hist, bin_edges

    for i in range(len(hist)):
        lower_bound = bin_edges[i]
        upper_bound = bin_edges[i+1]
        count = hist[i]

        num_chars = int((count / max_count) * max_character_per_line)
        if count > 0 and num_chars == 0:
            num_chars = 1

        bar = character * num_chars
        print(f"[{lower_bound: >3.0f}, {upper_bound: >3.0f}] : {bar}")
    return hist, bin_edges
```

```python
def where(mylist,myfunc):
    out= []
    for i, element in enumerate(mylist):
        if myfunc(element):
            out.append(i)
    return out
```

```python
def in_range(mymin,mymax):
    def testrange(x):
        return x<mymax and x>=mymin
    return testrange

F1=in_range(0,10)
F2=in_range(10,20)

print (F1(0), F1(1), F1(10), F1(15), F1(20))
print (F2(0), F2(1), F2(10), F2(15), F2(20))

print ("Number of Entries passing F1:", len(where(data,F1)))
print ("Number of Entries passing F2:", len(where(data,F2)))


def is_even():
    def test_even(x):
        return x % 2 == 0
    return test_even

def is_odd():
    def test_odd(x):
        return x % 2 != 0
    return test_odd

def greater_than(value):
    def test_greater(x):
        return x > value
    return test_greater

def less_than(value):
    def test_less(x):
        return x < value
    return test_less

def is_equal(value):
    def test_equal(x):
        return x == value
    return test_equal

def divisible_by(divisor):
    def test_divisible(x):

        if divisor == 0:
            return False
        return x % divisor == 0
    return test_divisible
```

```python
def generate_function(func,x_min,x_max,N=1000):
    out = list()

    sample_points = [x_min + i * (x_max - x_min) / 999 for i in range(1000)]
    y_values = [func(x) for x in sample_points]
    y_max = max(y_values)

    if y_max <= 0:
        print("Error")
        return []

    count = 0
    while count < N:

        test_x = random.uniform(x_min, x_max)
        p = random.uniform(0, y_max)
        if p <= func(test_x):
            out.append(test_x)
            count += 1
    return out
```

```python
def test_func(x,a=1,b=1):
    return abs(a*x+b)
    print("Generating data")
generated_data = generate_function(test_func, -10, 10, 1000)

print("\nGenerated data stats:")
print(f"  Number of points: {len(generated_data)}")
if generated_data:
    print(f"  Min value: {min(generated_data):.2f}")
    print(f"  Max value: {max(generated_data):.2f}")

print("\nHistogram of generated data:")
# Using the draw_histogram function from Exercise 4
draw_histogram(generated_data, n_bins=10, x_min=-10, x_max=10, character='*', max_character_per_line=50)
```
